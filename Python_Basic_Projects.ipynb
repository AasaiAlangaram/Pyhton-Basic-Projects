```python
"""
Binary Search Algorith for Descending ordered Array
"""

def binary_search_desc(des_array,start,end,target):
    
    middle = 0
    
    if(start <= end):
        
        if des_array[start] == target:
            return start
        if des_array[end] == target:
            return end
        
        middle = (start+end)//2
        #print(middle)
        
        if des_array[middle] == target:
            return middle
        
        if (des_array[middle] > target):
            return binary_search_desc(des_array,middle+1,end,target)
        
        if (des_array[middle] < target):
            return binary_search_desc(des_array,start,middle-1,target)
        
    return -1

des_array = [873, 870, 868, 865, 863, 860, 857, 854, 851, 848]

#print(len(des_array))
print('The value found at location :',binary_search_desc(des_array,0,len(des_array)-1,860))
```
